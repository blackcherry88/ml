```python
from scipy import signal
import numpy as np

def hoeffding (x, epsilon):
    y = 2.0 * np.exp(-2*epsilon*epsilon*x)
    return y
```

```python
steps = 100
#x = np.arange (0, 100, 0.01)
x = np.arange (0, steps, 0.01)

ps = [0.1, 0.3, 0.5]


ys = [hoeffding (x, p) for p in ps]

import matplotlib.pyplot as plt

plt.xlabel ('N')
plt.ylabel ('P')
latex1 = r'P\leq2e^{\left( -2\varepsilon ^{2}N\right)}'
plt.title (f"Hoeffding's Inequality: $ %s $" % latex1)

latext2 = r'\varepsilon'
for p, y in zip(ps, ys):
    plt.plot (x, y, label=f'$%s={p}$' % latext2)

plt.legend ()
plt.show ()
```
