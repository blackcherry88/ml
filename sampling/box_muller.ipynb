Box-Muller Algorithm
1. Generate $U1∼uniform(0,1)$ and $U2∼uniform(0,1)$ where $U1⊥U2$
2. Set $R=−2log(U1)$ and $\theta=2πU2$
3. Set $X=Rcos(\theta)$ and $Y=Rsin(\theta)$

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt


def p(x, mu=0, scale=1):
    return stats.norm.pdf(x, loc=mu, scale=scale)


np.random.seed(521)
U1 = np.random.uniform(size = 1000)
U2 = np.random.uniform(size = 1000)
R = np.sqrt(-2 * np.log(U1))
Theta = 2 * np.pi * U2
X = R * np.cos(Theta)
Y = R * np.sin(Theta)
```

```python
sns.displot(X, kind='kde', color='r')
plt.scatter(X, p(X))
plt.show()
```
